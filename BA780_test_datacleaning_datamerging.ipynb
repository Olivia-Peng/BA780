```python
import pandas as pd
kaggle = pd.read_excel("/content/Spotify and Youtube (test).xlsx")
#print(kaggle)
```

```python
import pandas as pd
wAPI = pd.read_excel("/content/Spotify_Youtube_wAPI (test).xlsx")
#print(wAPI)
```

```python
# Q1, what is the difference between ["Unnamed: 0.1"] & ["Unnamed: 0"] âœ…
  # figure out that there might be some duplicated value -> check

# shows that there really exist duplicated value
wAPI[wAPI.duplicated(subset=["Artist","Url_spotify","Track"])]

# drop the duplicated value
wAPI.drop_duplicates(subset=["Artist","Url_spotify","Track"],inplace = True)

# new wAPI without duplicated value
wAPI
```

```python
# Q2, if I want to join wAPI and kaggle data together, need to ensure that all the songs are in sync with each other âœ…

# set kaggle as true_data & set wAPI as study_data, find out if there any info in study_data that is out of the true_data range
difference = set(wAPI["Track"]).difference(kaggle["Track"])
print(difference)
```

```python
# Q3, drop of the ["Unnamed: 0.1"] from wAPI, it's a redundant information âœ…

wAPI.drop("Unnamed: 0.1",axis=1,inplace=True)
                      # make sure to set axis=1 (whole column)
```

```python
# just to make sure the 2 datasets are now joinable âœ…
wAPI.to_excel('wAPI.xlsx')
```

```python
# Q4, different columns between wAPI and kaggle âœ…
different_columns = set(wAPI.columns).difference(kaggle.columns)
print(different_columns)
```

```python
# Q5, add new three columns:{'Genres', 'Artist Popularity', 'Track Popularity'} to kaggle using outer join, create a whole new dataset = dataset âœ…

""" OUTER JOIN (return all of the rows and columns from both tables, doesn't matter if there is a match between the table)
f1.merge(f2,how="outer",suffixes=("_f1","_f2"),on="")
"""

common_columns = ['Artist', 'Url_spotify', 'Track', 'Album', 'Album_type', 'Uri',
       'Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo',
       'Duration_ms', 'Url_youtube', 'Title', 'Channel', 'Views', 'Likes',
       'Comments', 'Description', 'Licensed', 'official_video', 'Stream']

dataset = kaggle.merge(wAPI, on=common_columns  ,how="outer")
```
