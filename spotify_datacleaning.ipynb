```python
import pandas as pd
api = pd.read_csv("/content/API.csv")
```

```python
# Step 1-1: find out if there are any duplicated values, if yes -> drop them off

# Q: don't know why if I only set subset = "Uri", the result will be so weird

duplicated_value = api.duplicated(subset=["Artist","Track","Uri"])
api = api[~duplicated_value]
```

```python
api.info()
```

```python
api
```

```python
# step 1-2: drop off some non-use columns
api.drop(["Unnamed: 0.1","Url_spotify","Uri","Url_youtube","Description"],axis=1,inplace=True)
```

```python
# step 1-3: set the correct sequence label as an index, without any index label on it
api.set_index("Unnamed: 0",inplace=True)
api = api.rename_axis("")
```

```python
api.info()
```

```python
# step 2-1: GENRES  https://saturncloud.io/blog/how-to-split-text-in-a-column-into-multiple-rows-using-pandas/

api["Genres"] = api["Genres"].str.split(',')
api = api.explode("Genres")
        # used to deal with columns that contain lists or arrays
        # : to break apart these lists or arrays into individual elements, replicating the other column values as needed to maintain the DataFrame's integrity
```

```python
api
```

```python
# step 2-2: Datetime issue, extract release date: year, month, day
from datetime import datetime as dt
import pandas as pd

api['Release Date'] = pd.to_datetime(api['Release Date'], infer_datetime_format=True, errors='coerce')

api['Release year'] = api['Release Date'].dt.year
api['Release month'] = api['Release Date'].dt.month
api['Release day'] = api['Release Date'].dt.day
```

```python
# check use
print(max(api['Release year']))
print(min(api['Release year']))
print(max(api['Release month']))
print(min(api['Release month']))
print(max(api['Release day']))
print(min(api['Release day']))
```

```python
# drop off: based on "Stream","Title","Likes","Comments" columns
api.dropna(subset=["Stream","Title","Likes","Comments","Release Date"],inplace=True)
```

```python
"""
calculation:
sum_value('dance pop'+'pop'+'post-teen pop') / sum_count('dance pop'+'pop'+'post-teen pop')
"""
```

```python
import numpy as np
features_sum = api.groupby("Genres")["Danceability","Energy","Key","Loudness","Speechiness","Acousticness","Instrumentalness","Liveness","Valence","Tempo"].agg(np.sum)
features_count = api.groupby("Genres")["Danceability","Energy","Key","Loudness","Speechiness","Acousticness","Instrumentalness","Liveness","Valence","Tempo"].count()
```

```python
features_sum
```

```python
features_count
```

```python
features_sum.loc[["'dance pop'","'pop'","'post-teen pop'"]]
```

```python
features_count.loc[["'dance pop'","'pop'","'post-teen pop'"]]
```

```python
total_sum =  features_sum.loc[["'dance pop'","'pop'","'post-teen pop'"]].agg(np.sum)
print(total_sum)
print()
total_count = features_count.loc[["'dance pop'","'pop'","'post-teen pop'"]].agg(np.sum) - 3
print(total_count)
```

```python
mean_value = round(total_sum / total_count,4)
print(mean_value)
```

```python
api.loc[11890]["Danceability"].fillna(mean_value[0],inplace=True)
api.loc[11890]["Energy"].fillna(mean_value[1],inplace=True)
api.loc[11890]["Key"].fillna(mean_value[2],inplace=True)
api.loc[11890]["Loudness"].fillna(mean_value[3],inplace=True)
api.loc[11890]["Speechiness"].fillna(mean_value[4],inplace=True)
api.loc[11890]["Acousticness"].fillna(mean_value[5],inplace=True)
api.loc[11890]["Instrumentalness"].fillna(mean_value[6],inplace=True)
api.loc[11890]["Liveness"].fillna(mean_value[7],inplace=True)
api.loc[11890]["Valence"].fillna(mean_value[8],inplace=True)
api.loc[11890]["Tempo"].fillna(mean_value[9],inplace=True)
```
